defining values for solving transportation problem

```python
# Creates a list of all the supply nodes
warehouses = ["Office", "Factory"]#supply

# Creates a dictionary for the number of units of supply for each supply node
supply = {"Office": 10000, "Factory": 20000}#demand

# Creates a list of all demand nodes
projects = ["bihar", "chhatisgarh", "gujrat","karnataka","madhya","maharashtra","telegnana","utar"]

# Creates a dictionary for the number of units of demand for each demand node
demand = {
    "bihar": 3750,
    "chhatisgarh": 3750,
    "gujrat": 3750,
    "karnataka": 3750,
    "madhya": 3750,
    "maharashtra": 3750,
    "telegnana": 3750,
    "utar": 3750,
}

# Creates a list of costs of each transportation path
costs = {  # Projects
    'Factory':{"bihar":bihar_office,"chhatisgarh":chhatisgarh_office,"gujrat":gujrat_office,"karnataka":karnataka_office,"madhya":madhya_office,"maharashtra":maharashtra_office,"telegnana":telangana_office,"utar":uttar_office},  # A   warehouses
    'Office':{"bihar":bihar_chetak,"chhatisgarh":chhatisgarh_chetak,"gujrat":gujrat_chetak,"karnataka":karnataka_chetak,"madhya":madhya_chetak,"maharashtra":maharashtra_chetak,"telegnana":telangana_chetak,"utar":uttar_chetak}  # B
}

# The cost data is made into a dictionary
```

setting problem

```python
from pulp import *
prob = LpProblem("Transportation", LpMinimize)
```

```python
routes =[(i,j) for i in warehouses for j in projects]
```

defining decision variable

```python
amount_vars = LpVariable.dicts("Amountship",(warehouses,projects),0)
```

defining objective function

```python
prob += lpSum(amount_vars[i][j]*costs[i][j] for (i,j) in routes)
```

constraint

```python
for j in projects:
    prob += lpSum(amount_vars[i][j] for i in warehouses) == demand[j]

for i in warehouses:
    prob += lpSum(amount_vars[i][j] for j in projects) <= supply[i]
```

to check if the problem is solved or not

```python
prob.solve()
```

check if optimal solution is attained or not

```python
print("Status:", LpStatus[prob.status])
```

value in every cell of transportation table

```python
for v in prob.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
```
